```python
import pandas
```

```python
import geopy
```

```python
df = pandas.read_csv('pandas_data/supermarkets.csv')
df
```

```python
from geopy.geocoders import Nominatim
nom = Nominatim()
n = nom.geocode('3995 23rd St, San Francisco, California 94114')
n
```

```python
print(n.latitude)
print(n.longitude)
```

```python
# Returns None if address doesn't exist
m = nom.geocode('735 Dolores St, San Francisco, California 94119')
print(m)
```

```python
# Modify address column
df['Address'] = df['Address'] + ', ' + df['City'] + ', ' + df['State'] + ', ' + df['Country']
df
```

```python
# NO NEED FOR LOOPING
df['Coordinates'] = df['Address'].apply(nom.geocode)
df
```

```python
print(df.Coordinates[0])
print(df.Coordinates[0].latitude)
```

```python
df['Latitude'] = df['Coordinates'].apply(lambda x: x.latitude if x is not None else None)
df['Longitude'] = df['Coordinates'].apply(lambda x: x.longitude if x is not None else None)
df
```
